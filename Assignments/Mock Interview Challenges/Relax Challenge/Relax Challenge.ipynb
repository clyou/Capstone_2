```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
users_df = pd.read_csv("takehome_users.csv", encoding='ISO-8859-1')

engagement_df = pd.read_csv("takehome_user_engagement.csv", encoding='ISO-8859-1')
```

```python
users_df.head() 
```

```python
engagement_df.head()
```

```python
engagement_df['adopted_user'] = 0 #Creating new column for 'adopted_user'
engagement_df.head()
```

```python
len(engagement_df)
```

```python
len(users_df)
```

```python
users_df.rename(columns = {'object_id':'user_id'}, inplace = True) #Do this prior to merging the 2 dataframes
users_df.head()
```

```python
data = pd.merge(users_df, engagement_df, on='user_id') #Merging the two dataframes, should have same length as engagement_df
len(data)
```

```python
data.head()
```

```python
#The Unix time needs to be converted to Datetime

data['last_session_creation_time'] = pd.to_datetime(data['last_session_creation_time'], unit='s')

data.head()
```

```python
data = data.sort_values(['user_id', 'time_stamp'], ascending = True)
data.head()
```

```python
type(data['time_stamp'][7])
```

```python
#Need to convert the string 'time_stamp' to a Datetime

data['time_stamp'] = pd.to_datetime(data['time_stamp'])

data.head()
```

```python
#This variable gives the number of days between every other observation. An aopted user is defined as one who logs
# in three times in a 7 day period. If sorted, then what is relevant is the endpoints. 

data['delta'] = data['time_stamp'] - data['time_stamp'].shift(-2)

data.head()
```

```python
#Converting the delta into the absolute value of days

data.delta.fillna(value=0, inplace=True)
data['delta'] = (data['delta'] / np.timedelta64(1, 'D')).astype(int)
data.delta = np.abs(data.delta)
data.head()
```

```python
#This variable tells us if the user_id is the same between two observations, 0 = same, >0 = different

data['same_user'] = np.abs(data['user_id'] - data['user_id'].shift(-2))
data.same_user.fillna(value=1, inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
#One assumes that if there is no invite, the user_id of the inviter is 0. 

data.invited_by_user_id.fillna(value=0, inplace = True)
data.isnull().sum()
```

```python
data.head()
```

```python
#To determine who is an adopted_user and who isn't, the user_ids must be the same and the delta must be less than 7 days 

data.loc[((data['delta'] <= 7) & (data['same_user'] == 0)), 'adopted_user'] = 1

data.head()
```

```python
#Determine what number of entries is consistent with being an adopted_user

data['adopted_user'].sum()
```

```python
data.describe().T
```

```python
data['same_user'].unique()
```

```python
# Number of unique user_ids
data['user_id'].nunique()
```

```python
# Organizational IDs
print(data['org_id'].unique())
```

```python
data['creation_source'].unique()
```

```python
# It looks like out of 8823 user_ids, 2230 are inviters

data['invited_by_user_id'].nunique()
```

```python
# We want to know who is a recommender and if that affects the quality of being an adopted user

# This creates a list of inviter user ids
recommend_set = list(data['invited_by_user_id'].unique()) 
print(len(recommend_set))
```

```python
#This step determines who is a recommender and who isn't.

data['recommender'] = 0
data['user_id2'] = data['user_id'] * 1.0
data['rec'] = np.isin(data['user_id2'], recommend_set)

data.loc[data['rec'] == False, 'recommender'] = 0
data.loc[data['rec'] == True, 'recommender'] = 1


data['recommender'].sum()
```

```python
data['recommended'] = 0
data.loc[data['invited_by_user_id'] > 0, 'recommended'] = 1
```

```python
data['recommended'].sum()
```

```python
data['visited'].count() #Superfluous column
```

```python
data.head()
```

```python
# Getting rid of unnecessary coumns

df = data.drop(['name', 
                'email', 
                'user_id2', 
                'same_user', 
                'rec', 
                'visited', 
                'invited_by_user_id', 
                'delta'], axis = 1).sort_index()
df.head()
```

```python
y = data['adopted_user'] #PREDICTOR / RESPONSE VARIABLE
```

```python
creation_dummy = pd.get_dummies(df['creation_source'], prefix = 'Is', drop_first=True)
creation_dummy.describe().T

#GUEST_INVITE is the base case
```

```python
# A logistic regression model is used because the response variable is binary. 


X1 = sm.add_constant(creation_dummy)

logit = sm.Logit(y, X1)

result = logit.fit()
print(result.summary())
```

```python
# odds ratios only
print(np.exp(result.params))
```

```python
org_dummy = pd.get_dummies(df['org_id'], prefix = 'Is', drop_first=True)
org_dummy.describe().T
```

```python
X2 = sm.add_constant(org_dummy)

logit = sm.Logit(y, X2)

result = logit.fit()
print(result.summary())
```

```python
# odds ratios only
print(np.exp(result.params))
```

```python
X3 = df[['opted_in_to_mailing_list', 'enabled_for_marketing_drip', 'recommender', 'recommended']]
X3.head()
```

```python
X3 = sm.add_constant(X3)

logit = sm.Logit(y, X3)

result = logit.fit()
print(result.summary())
```

```python
# odds ratios only
print(np.exp(result.params))
```
